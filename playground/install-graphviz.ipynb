```python
from graphviz import Digraph

dot = Digraph(comment='Tree')
heap = "ABCDEFG"

for i in heap:
    dot.node(str(i), str(i))
def draw(heap):
    for parent in range(int(len(heap)/2)+1):
        left = parent*2+1
        right = parent*2+2
        if left < len(heap):
            dot.edge(str(heap[parent]), str(heap[left]))
        if right < len(heap):
            dot.edge(str(heap[parent]), str(heap[right]))

draw(heap)
dot
```
